```python
#### Count words in a text file
text=open("auto_close_sample_124796855.xml.txt","r")
d=dict()
for line in text:
    line=line.strip()
    line=line.lower()
    words=line.split(" ")
    for word in words:
        word=word.strip()
        print("Key is ["+word+"]")
        if word in d:
            d[word]=d[word]+1
        else:
            d[word]=1

print("<ticketlist>",":",d["<ticketlist>"])
#,":",d["<ticketList>"])
for key in d.keys():
    print(key,":",d[key])
```
